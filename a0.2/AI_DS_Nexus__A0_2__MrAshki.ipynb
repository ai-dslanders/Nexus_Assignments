### Task 3 ‚Äî Load ride log


```python
# üëâ # TODO: read "rides.csv" into df
# (columns: date,temp,rides,weekday)
import pandas as pd


df = pd.read_csv("rides.csv")

print(df.head())

```

### Task 4 ‚Äî Weekday averages

```python
# üëâ # TODO: compute and print mean rides per weekday
# Task 4
weekday_avg = df.groupby('weekday')[['rides']].mean().reset_index()
print(weekday_avg)
```

### Task 5 ‚Äî Scatter plot

```python
# üëâ # TODO: scatter-plot temperature (x) vs rides (y) from df
import matplotlib.pyplot as plt


# Task 5: ÿß€åÿ¨ÿßÿØ ŸÜŸÖŸàÿØÿßÿ± Ÿæÿ±ÿß⁄©ŸÜÿØ⁄Ø€å
plt.scatter(df['temp'], df['rides'], alpha=0.5, c='blue')

# ÿßÿ∂ÿßŸÅŸá ⁄©ÿ±ÿØŸÜ ÿ¨ÿ≤ÿ¶€åÿßÿ™ ÿ®ÿ±ÿß€å ÿØÿ±⁄© ÿ®Ÿáÿ™ÿ± (ÿßÿÆÿ™€åÿßÿ±€å ÿßŸÖÿß ÿ≠ÿ±ŸÅŸá‚Äåÿß€å)
plt.title('Relationship between Temperature and Rides')
plt.xlabel('Temperature (¬∞C)')
plt.ylabel('Number of Rides')
plt.show()
```
